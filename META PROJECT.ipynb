```python
meta["Average 3"]=(meta["Close"]+meta["Close"].shift(1)+meta["Close"].shift(2))/3
meta.head()
```

```python
meta["MA10"]=meta["Close"].rolling(10).mean()
meta["MA50"]=meta["Close"].rolling(50).mean()
```

```python
import numpy as np
from scipy.stats import norm
```

```python
meta["LogReturn"] = np.log(meta["Close"].shift(-1)) - np.log(meta["Close"])# calculating log daily return
```

```python
mu=meta["LogReturn"].mean()
sigma=meta["LogReturn"].std(ddof=1)

density=pd.DataFrame()
density["x"]=np.arange(meta["LogReturn"].min()-0.01,meta["LogReturn"].max()+0.01,0.001)
density["PDF"]=norm.pdf(density["x"],mu,sigma)

meta["LogReturn"].hist(bins=50,figsize=(15,8))
plt.plot(density["x"],density["PDF"],color="red")
plt.show()

#Plot the histrogram to show the distribution of long return of META stocks
#We can see it is very close to a norml distribution
```

```python
#calculating the probability of stock price will drop over certain percentage in a day
```

```python
#probability that the stock price of meta will drop over 5% in day
prob_return=norm.cdf(-0.05,mu,sigma)
print("The probability of stock price will drop over 5% in a day is",prob_return)
```

```python
#calculating the probability of stock price will drop over certain percentage in a year
```

```python
#drop over 40% in 220 days
mu220=220*mu
sigma220=(220**0.5)*sigma
print("The probability of dropping over 40% in 220 days is", norm.cdf(-0.4,mu220,sigma220))
```

```python
# calculating the Var(Value of Risk)
```

```python
var=norm.ppf(0.05,mu,sigma)#ppf meaning is Percent-Point Function
print("Single day value at risk",var)
```

```python
# Quatile
# 5% quatile
print("5% quantile",norm.ppf(0.05,mu,sigma))#ppf meaning is Percent-Point Function 
# 95% quatile
print("95% quantile",norm.ppf(0.95,mu,sigma))
```

```python
# Estimate the average stock return with 90% Confidence Interval
```

```python
sample_size = meta['LogReturn'].shape[0]
sample_mean =meta['LogReturn'].mean()
sample_std =meta['LogReturn'].std(ddof=1) / sample_size**0.5

# left and right quantile
z_left = norm.ppf(0.05)
z_right = norm.ppf(0.95)

# upper and lower bound
interval_left = sample_mean+z_left *sample_std
interval_right = sample_mean+z_right*sample_std

print("Sample mean is ",sample_mean)
print("90% confidence interval is")
print(interval_left,interval_right,)

#we can see the output is positive there fore average stock return is positive
```
